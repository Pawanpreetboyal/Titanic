## Adding Fake Employment Data to Student Survey Dataset

### Why We Created Fake Data?
Since no real dataset was available to merge with our **Cleaned Student Survey Data**, I created a **fake dataset** using Chatgpt containing employment statistics based on the **Program** field.
This dataset helps analyze:
- **Employment rates (%)** for graduates of each program.
- **Industry placements** (e.g., BITMAN graduates go into Tech, Babi students into Analytics, HRMG into Human Resources ).
- **Average salary after graduation** for different programs.


```python
# Loading the fake employment dataset
employment_data = pd.read_csv("C:/Users/Hp/Downloads/standardized_fake_employment_data.csv")
```

### **Type of Merge Used**
We used a **left join (`how="left"`)** because:
- It ensures **all students remain in the dataset**, even if their program doesn't have employment data.
- If employment data is missing for a program, the new columns (`Employment_Rate`, `Industry_Placement`, `Avg_Salary_After_Graduation`) will contain **NaN values**.
- This approach keeps the dataset intact while adding valuable employment insights.


### **Handling Missing Values (`NaN`)**
Some students may have **NaN values** in employment-related columns. This happens if:
- Their program was **not included** in the fake employment dataset.
- The dataset lacks **employment statistics** for that specific program.

We can handle missing values by:
- **Filling NaNs with "Unknown"** for categorical fields like `Industry_Placement`.
- **Replacing NaNs with median values** for numerical fields like `Avg_Salary_After_Graduation`.



```python
# Merged using "Program" as the key
merged_data = pd.merge(cleaned_data, employment_data, on="Program", how="left")
```

```python
# Define the path to Downloads
downloads_path = os.path.expanduser("~/Downloads/asgnmerged_student_employment_data.csv")

```

### **Final Dataset Saved to downlaods**

```python
# Save the merged dataset to Downloads
merged_data.to_csv(downloads_path, index=False)

```
